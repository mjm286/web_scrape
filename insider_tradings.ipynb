```python
# Imports
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from bs4 import BeautifulSoup
import pandas as pd
from time import sleep
import numpy as np
import itertools
from IPython.core.display import display, HTML
display(HTML('<style>.container {width:100% !important; }</style>'))
```

```python

def finviz_scrape(symbol):
    retry_count = itertools.count()
    print(symbol)
    driver=None
    while True: # initate search
        try:
            sleep(.25)
            driver = webdriver.Chrome() # open chrome
            driver.get('https://finviz.com/')
            driver.find_element_by_xpath('//*[@id="search"]/div/form/input').send_keys(symbol) # assign symbol to search box
            driver.find_element_by_xpath('//*[@id="search"]/div/form/span').click() # execute search query

            soup = BeautifulSoup(driver.page_source, "lxml") # obtain page source using BeautifulSoup package
            table = soup.findAll("table", {"class":"body-table"})[0] # identify insider trading table


            info = [[td.get_text() for td in row.find_all("td")] # scrape rows in table
                    for row in table.find_all("tr")]

            df=pd.DataFrame(info)
            df['symbol'] = symbol
            driver.quit() # close chrome
            return df
            break
        except: # if stock has no recorded insider trading
            if next(retry_count) <= 0: # search again, just to be sure
                print('not there, try again')
                driver.quit()
            else: # breakout
                driver.quit()
                print('not found')
                break
```

```python
symbol='airt' # Air T Corporation
inside_trading =finviz_scrape(symbol)
```
